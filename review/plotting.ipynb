```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
toronto = pd.read_csv("../data/Toronto.csv", parse_dates=["Date"], index_col=False) # prevents errors due to malformed data
#toronto.set_index('Date', inplace=True)
montreal = pd.read_csv("../data/Montreal.csv", parse_dates=["Date"],index_col=False)
recife = pd.read_csv("../data/Recife.csv", parse_dates=["Date"],index_col=False)
```

```python
toronto_monthly = toronto.groupby([toronto['Date'].dt.year, toronto['Date'].dt.month]).mean()
montreal_monthly = montreal.groupby([montreal['Date'].dt.year, montreal['Date'].dt.month]).mean()
recife_monthly = recife.groupby([recife['Date'].dt.year, recife['Date'].dt.month]).mean()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
plt.xlim(1, 12)
ax.plot(toronto_monthly.loc[2013]["Max Temperature"], label='Toronto')
ax.plot(montreal_monthly.loc[2013]["Max Temperature"], label='Montreal')
ax.plot(recife_monthly.loc[2013]["Max Temperature"], label='Recife')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature')
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.grid()
ax.legend()
```
