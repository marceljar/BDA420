#### Question 1 (0.5 mark)
Retrieve the information from the *bda420.csv* file, and store it in a pandas dataframe. Following, print a list with the names of all fields (columns).

```python
df = pd.read_csv("/data/bda420.csv")
print(df.columns.values)
```

#### Question 2 (0.5 mark)
Obtain the following statistics about the *bda420.csv* data, and print them:
- The highest GPA
- The lowest GPA
- The average GPA

Note that your output for the first two statistics should contain the name of the student. For Example: *The student with highest gpa was John Smith with a GPA of 3.95*

```python
max = df["GPA"].max()
min = df["GPA"].min()
avg = df["GPA"].mean()

print(f"The highest GPA was {max:.2f}, the minimum was {min:.2f}, and the aerage was {avg:.2f}.")
```

#### Question 3 (1.0 marks)
Display the names and GPAs of all students with a GPA equal or higher than 3.5 in ascending order. Your answer should print only the students names and their GPAs.

```python
honour_list = df[df['GPA'] >= 3.5]
sorted_honour_list = honour_list.sort_values(by='GPA', ascending=False)

print("The students in our honour list are:")
for index, row in sorted_honour_list.iterrows():
    print(f" {row['Full_Name']}: {row['GPA']}")
```

#### Question 4 (1.0 marks)

Plot an histogram with the grade distribution of all students. Your histogram should have 6 bins and proper axis labels.

```python
plt.hist(df['GPA'], bins=6, color='red', edgecolor='black')
plt.title('GPA Distribution')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```
