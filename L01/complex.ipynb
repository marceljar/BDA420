```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/data/simple.csv")

df["revenue"] = df["units_sold"] * df["unit_price"]
```

```python
# Revenue by category (bar chart)
rev_by_cat = df.groupby("category")["revenue"].sum().sort_values(ascending=False)

rev_by_cat.plot(kind="bar")
plt.title("Total Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
# Top 10 days by revenue (bar chart)
top_days = df.sort_values("revenue", ascending=False).head(10)

top_days.plot(x="day", y="revenue", kind="bar")
plt.title("Top 10 Days by Revenue")
plt.xlabel("Day")
plt.ylabel("Revenue")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Revenue trend over time (line chart)
df_sorted = df.sort_values("day")

df_sorted.plot(x="day", y="revenue", kind="line", marker="o")
plt.title("Revenue Over Time")
plt.xlabel("Day")
plt.ylabel("Revenue")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
