```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
        .appName("Basic CSV Operations") \
        .getOrCreate()
```

```python
# Adjust path if the file is in a subfolder
csv_path = "/data/sales.csv"
df = spark.read.csv(csv_path, header=True, inferSchema=True)
df.show(5)
```

```python
df.printSchema()
```

```python
high_value_orders = df.filter(df.amount > 100)
high_value_orders.show()
```
